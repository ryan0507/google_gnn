```python
import seaborn
import json
```

```python
epoch = []
loss = []
opa = []
with open('./stats_val.json', 'r') as f:
    for line in f:
        data = json.loads(line)
        epoch.append(data['epoch'])
        loss.append(data['loss'])
        opa.append(data['opa'])
```

```python
ckpt_opa = []
for idx in range(99,len(opa)-1, 100):
    ckpt_opa.append([idx, opa[idx]])
```

```python
sorted(ckpt_opa, lambda x:(-x[1],x[0]))
```
